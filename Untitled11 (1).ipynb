2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for discrete uniform distribution: values from 1 to 6
low = 1
high = 7

# Define the distribution: randint is inclusive of low and high in this usage
x = np.arange(low, high + 1)  # [1, 2, 3, 4, 5, 6]
pmf_values = [1 / (high - low + 1)] * len(x)  # Equal probabilities

# Plot the PMF
plt.figure(figsize=(8, 4))
plt.stem(x, pmf_values, basefmt=" ")
plt.title('PMF of Discrete Uniform Distribution (1 to 7)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.xticks(x)
plt.grid(True)
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.


```python
def bernoulli_pmf(x, p):

    if x not in [0, 1]:
        return 0  # Only 0 and 1 are valid outcomes
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    return p if x == 1 else 1 - p

# Example usage:
print("P(X=1):", bernoulli_pmf(1, 0.7))  # P(success)
print("P(X=0):", bernoulli_pmf(0, 0.7))  # P(failure)
```

6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the discrete uniform distribution
low = 1
high = 6

# Discrete outcomes: 1 through 6
x = np.arange(low, high + 1)

# PMF: All outcomes are equally likely
pmf = np.full_like(x, 1 / len(x), dtype=np.float64)

# CDF: Cumulative sum of PMF values
cdf = np.cumsum(pmf)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', label='CDF', color='blue', linewidth=2)
plt.scatter(x, cdf, color='red')  # dots at steps
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.yticks(np.linspace(0, 1, len(x)+1))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores and plot them.

    Parameters:
    data : list or numpy array
        A numerical dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='teal', label='Z-Scores')
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axhline(2, color='red', linestyle='--', label='Z = ¬±2')
    plt.axhline(-2, color='red', linestyle='--')
    plt.title('Z-Scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage:
sample_data = [10, 12, 13, 9, 15, 18, 21, 14, 16, 17]
z = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", z)
```

14.5 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    x : float
        Data point
    mean : float
        Mean of the dataset
    std : float
        Standard deviation of the dataset

    Returns:
    float : Z-score
    """
    return (x - mean) / std

def plot_z_score(z):
    """
    Plot the standard normal distribution and mark the Z-score.

    Parameters:
    z : float
        Z-score to plot
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, color='blue', label='Standard Normal PDF')

    # Shade the area under the curve up to the Z-score
    x_fill = np.linspace(-4, z, 500)
    y_fill = norm.pdf(x_fill)
    plt.fill_between(x_fill, y_fill, alpha=0.3, color='green', label=f'Area ‚â§ Z = {z:.2f}')

    plt.axvline(z, color='red', linestyle='--', linewidth=2, label=f'Z = {z:.2f}')
    plt.title('Z-score Visualization on Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
x = 78       # data point
mean = 70    # sample/population mean
std = 5      # sample/population standard deviation

z_score = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x} is: {z_score:.2f}")

# Plot Z-score on the normal distribution
plot_z_score(z_score)
```

15.5 Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, mu_0, alpha=0.05):
    """
    Perform Z-test for population mean.

    Parameters:
    sample_data : list or np.array
        The sample data
    mu_0 : float
        The hypothesized population mean (null hypothesis)
    alpha : float
        Significance level (default is 0.05 for a 95% confidence level)

    Returns:
    z : float
        Computed Z-statistic
    p_value : float
        Two-tailed p-value
    """
    sample_data = np.array(sample_data)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # sample standard deviation

    # Compute Z-statistic
    z = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

    # Compute two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.3f}")
    print(f"P-value: {p_value:.4f}")

    # Hypothesis decision
    if p_value < alpha:
        print("‚ùå Reject the null hypothesis (significant difference).")
    else:
        print("‚úÖ Fail to reject the null hypothesis (no significant difference).")

    return z, p_value

# Example sample dataset
data = [102, 98, 101, 100, 99, 97, 103, 104, 99, 100]

# Hypothesized population mean (e.g., from historical data)
mu_0 = 100

# Run the Z-test
z_test(data, mu_0)
```

16.Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    data1, data2 : list or array
        The two datasets to compare
    """
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")

    # Combine data for plotting
    combined_data = [data1, data2]

    # Boxplot
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.boxplot(data=combined_data)
    plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
    plt.title('Boxplot Comparison')
    plt.ylabel('Values')

    # Histogram
    plt.subplot(1, 2, 2)
    sns.histplot(data1, color='blue', kde=True, label='Dataset 1', alpha=0.5)
    sns.histplot(data2, color='orange', kde=True, label='Dataset 2', alpha=0.5)
    plt.title('Histogram Comparison')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Example datasets
data1 = np.random.normal(50, 5, 100)   # mean=50, std=5
data2 = np.random.normal(50, 15, 100)  # same mean, larger std

compare_standard_deviations(data1, data2)
```

17.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution5

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a dataset.

    Parameters:
    data : list or array
        Numerical dataset

    Returns:
    tuple : (range, IQR)
    """
    data_range = np.max(data) - np.min(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    print(f"Range: {data_range:.2f}")
    print(f"IQR (Interquartile Range): {iqr:.2f}")
    return data_range, iqr

# Step 1: Generate normally distributed data
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=200)  # mean=100, std=15

# Step 2: Calculate and print range and IQR
calculate_range_and_iqr(data)

# Step 3: (Optional) Plot boxplot for visualization
plt.figure(figsize=(6, 4))
plt.boxplot(data, vert=False)
plt.title('Boxplot of Normally Distributed Data')
plt.xlabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def z_score_normalize(data):
    """
    Normalize a dataset using Z-score normalization.

    Parameters:
    data : array-like
        Input dataset

    Returns:
    np.ndarray : Z-score normalized dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_normalized = (data - mean) / std

    print(f"Original Mean: {mean:.2f}")
    print(f"Original Std Dev: {std:.2f}")
    print(f"Normalized Mean: {np.mean(z_normalized):.2f}")
    print(f"Normalized Std Dev: {np.std(z_normalized, ddof=1):.2f}")

    return z_normalized

# Step 1: Generate a normally distributed dataset
np.random.seed(42)
data = np.random.normal(loc=60, scale=10, size=200)  # mean=60, std=10

# Step 2: Apply Z-score normalization
z_data = z_score_normalize(data)

# Step 3: Visualize original vs normalized data
plt.figure(figsize=(12, 5))

# Histogram Comparison
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, color='skyblue', label='Original Data')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(z_data, kde=True, color='orange', label='Z-Score Normalized')
plt.title('Z-Score Normalized Distribution')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()

# Optional: Boxplot comparison
plt.figure(figsize=(6, 4))
plt.boxplot([data, z_data], labels=["Original", "Z-Normalized"])
plt.title("Boxplot Comparison")
plt.grid(True)
plt.tight_layout()
plt.show()
```
